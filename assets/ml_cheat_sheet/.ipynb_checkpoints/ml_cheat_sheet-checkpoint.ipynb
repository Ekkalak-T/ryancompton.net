```python

import sklearn
from sklearn.linear_model import LogisticRegression

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
```

```python
def clean(x):
    return x.strip().split()

def q1():
    dfx = pd.read_csv('q1x.dat', sep=';',
                      header=0, names=['durh'])
    dfx['x0'] = dfx['durh'].map(lambda x:float(clean(x)[0]))
    dfx['x1'] = dfx['durh'].map(lambda x:float(clean(x)[1]))
    dfx = dfx[['x0','x1']]
    
    dfy = pd.read_csv('q1y.dat',header=0,names=['y'])
    df = pd.concat([dfx,dfy],axis=1)
    
    logger = LogisticRegression(penalty='l1', tol=1e-5)
    X = df[['x0','x1']].as_matrix()
    y = df['y'].as_matrix()
    logger.fit(X,y)
    
    return df, logger
df,logger = q1()

def plotter(logger, df, npts=5000):
    xs = []
    ys = []
    cs = []
    for _ in range(npts):
        x0spr = df['x0'].max()-df['x0'].min()
        x1spr = df['x1'].max()-df['x1'].min()
        x = np.random.rand()*x0spr + df['x0'].min()
        y = np.random.rand()*x1spr + df['x1'].min()
        xs.append(x)
        ys.append(y)
        cs.append(logger.predict([x,y]))
    plt.scatter(xs,ys,c=cs, alpha=.25)
    plt.hold(True)
    plt.scatter(df['x0'],df['x1'],
                c=df['y'].map(lambda x:'r' if x else 'y'), 
                linewidth=0,s=20)
    plt.show()
    return


plotter(logger,df)
```

```python
import sklearn.linear_model
import sklearn.svm
import sklearn.tree
import sklearn.lda
import sklearn.naive_bayes
import sklearn.neighbors

import sklearn.datasets
X, y = sklearn.datasets.make_moons(n_samples=1500, noise=.15, random_state=None)

#straight
#classifier = sklearn.linear_model.

#curvy
#classifier = sklearn.svm.SVC()

#looks overfit
#classifier = sklearn.tree.DecisionTreeClassifier()

#lda is linear, straigt line
#classifier = sklearn.lda.LDA(n_components=2)

#bayes - straight
#classifier = sklearn.naive_bayes.GaussianNB()

#k nearest neigh
classifier = sklearn.neighbors.RadiusNeighborsClassifier()

classifier.fit(X,y)

dfX = pd.DataFrame(X,columns=['x0','x1'])
dfy = pd.DataFrame(y,columns=['y'])
df = pd.concat([dfX,dfy],axis=1)
plotter(classifier,df,npts=10000)
```
